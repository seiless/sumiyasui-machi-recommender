```python
from pathlib import Path
import requests
import pandas as pd
import os
from dotenv import load_dotenv
env_path = Path('..') / '.env'
load_dotenv(dotenv_path=env_path)
```

```python
URL = "https://www.reinfolib.mlit.go.jp/ex-api/external/XIT001"
PREFECTURE_URL = "https://www.reinfolib.mlit.go.jp/ex-api/external/XIT002"
KEY = os.getenv("REINFOLIB_CLIENT_API")
```

```python
headers = {
    "Ocp-Apim-Subscription-Key":KEY
}
params = {"year": 2024,
          "area":"01"}
response = requests.get(URL, headers=headers,params=params)
print(response.status_code)
```

```python
response_prefecture = requests.get(PREFECTURE_URL,headers=headers,params={"area":"01"})
print(response_prefecture.status_code)
```
