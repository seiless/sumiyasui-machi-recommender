```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import logging
from urllib.parse import urljoin
```

```python
HEADERS = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Accept-Language": "ja-JP,ja;q=0.9",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
CRILINAL_TARGETS = {
    "東京都": "https://www.keishicho.metro.tokyo.lg.jp/about_mpd/jokyo_tokei/jokyo/ninchikensu.html"  # noqa: E501
}

SINGLE_TARGET_URL = CRILINAL_TARGETS["東京都"]
```

```python
# 1. HTTPリクエストを送信
response = requests.get(SINGLE_TARGET_URL, headers=HEADERS)
response.raise_for_status()

# 2. BeautifulSoupでHTMLをパース (lxml 파서를 명시적으로 지정!)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
links = soup.find_all('a', class_='csv')
```

```python
for address in links:
    print(urljoin(SINGLE_TARGET_URL, address.get('href')))
```
