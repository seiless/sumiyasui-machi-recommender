```python
listings = soup.find_all('div', class_='property property--highlight js-property js-cassetLink')
properties_data = []
for listing in listings:
    # --- 각 항목을 추출하기 전, 기본값을 설정 ---
    title, address, transport_str, rent, admin_fee, deposit, key_money, layout, area, building_type, age = ['N/A'] * 11
    
    try:
        # --- 제목 추출 (최상위 컨테이너에만 있음) ---
        title_element = listing.find('a', class_='js-cassetLinkHref')
        if title_element:
            title = title_element.text.strip()
            
        # [사용자님 발견 2] 상세 정보가 담긴 'detailbox' 컨테이너 찾기
        detailbox = listing.find('div', class_='detailbox')
        if detailbox:
            # --- 'detailbox' 내부의 정보를 파싱 ---
            
            # 1. 가격/면적 등이 담긴 메인 테이블
            property_table = detailbox.find('div', class_='detailbox-property')
            if property_table:
                cols = property_table.find_all('td')
                if len(cols):
                    # 월세, 관리비
                    rent_div = cols[0].find('div', class_='detailbox-property-point')
                    if rent_div: rent = rent_div.text.strip()
                    admin_fee_div = cols[0].find('div', string=lambda t: t and '管理費' in t)
                    if admin_fee_div: admin_fee = admin_fee_div.text.strip()

                    # 보증금, 사례금
                    for div in cols[1].find_all('div'):
                    # div 안에 'square_pct' 클래스를 가진 span이 있는지 먼저 확인합니다.
                        span = div.find('span', class_='square_pct')
                        if span:
                            # span이 존재할 경우에만, 그 텍스트를 확인하여 값을 할당합니다.
                            label_text = span.text.strip()
                            if '敷' in label_text:
                                # 전체 텍스트에서 라벨 텍스트를 제거하여 순수한 값만 추출합니다.
                                deposit = div.text.strip().replace(label_text, '').strip()
                            elif '礼' in label_text:
                                key_money = div.text.strip().replace(label_text, '').strip()
                    
                    # 방 구조, 면적
                    madori_divs = cols[2].find_all('div')
                    if len(madori_divs) >= 2:
                        layout = madori_divs[0].text.strip()
                        area = madori_divs[1].text.strip()

                    # 건물 종류, 연식
                    type_divs = cols[3].find_all('div')
                    if len(type_divs) >= 2:
                        building_type = type_divs[0].text.strip()
                        age = type_divs[1].text.strip()

                    # 주소
                    address = cols[4].text.strip()

            # 2. 교통 정보 추출
            transport_box = detailbox.find('div', class_='detailnote-box')
            if transport_box:
                transport_divs = transport_box.find_all('div')
                transport_list = [t.text.strip() for t in transport_divs if t.text.strip() and 'お電話' not in t.text]
                transport_str = ' | '.join(transport_list)

        # 추출한 정보를 딕셔너리로 정리
        property_info = {
            '物件名': title, '住所': address, '交通': transport_str, '賃料': rent,
            '管理費': admin_fee, '敷金': deposit, '礼金': key_money,
            '間取り': layout, '専有面積': area, '種別': building_type, '築年数': age
        }
        properties_data.append(property_info)
        
    except Exception as e:
        print(f"データ抽出中にエラーが発生: {e}。この物件({title})はスキップします。")
        continue

# 최종 결과를 DataFrame으로 변환
df = pd.DataFrame(properties_data)

# 결과 확인
print(f"\n合計 {len(df)}件の物件情報を収集しました。")
df
```

```python
rooms = soup.find_all('div', class_='property_group')

for room in rooms:
    try:
        # [수정] 새로운 내부 구조에 맞춘 선택자 변경
        title = room.find('h2', class_='property_name').text.strip()
        
        # 주소, 교통, 건물 정보 추출
        address_element = room.find('div', class_='property_address')
        address = address_element.text.strip() if address_element else 'N/A'
        
        transport_elements = room.find_all('div', class_='property_access-title')
        transport = [t.text.strip() for t in transport_elements]
        
        age_element = room.find('div', class_='property_age')
        age_info = age_element.text.replace('\n', ' ').strip() if age_element else 'N/A'

        # 가격 및 상세 정보 추출
        price_element = room.find('div', class_='property_price')
        rent = price_element.find('div', class_='property_data-title').text.strip() if price_element else 'N/A'
        
        fee_element = room.find('div', class_='property_fee')
        if fee_element:
            admin_fee_span = fee_element.find('span', string='管理費・共益費')
            admin_fee = admin_fee_span.find_next_sibling('p').text.strip() if admin_fee_span else 'N/A'

            shikikin_span = fee_element.find('span', string='敷金')
            deposit = shikikin_span.find_next_sibling('p').text.strip() if shikikin_span else 'N/A'
            
            reikin_span = fee_element.find('span', string='礼金')
            key_money = reikin_span.find_next_sibling('p').text.strip() if reikin_span else 'N/A'
        else:
             admin_fee, deposit, key_money = 'N/A', 'N/A', 'N/A'
        
        madori_element = room.find('div', class_='property_madori')
        layout = madori_element.find('p').text.strip() if madori_element else 'N/A'
        
        menseki_element = room.find('div', class_='property_menseki')
        area = menseki_element.find('p').text.strip() if menseki_element else 'N/A'

        property_info = {
            '物件名': title, '住所': address, '交通': ' | '.join(transport),
            '建物情報': age_info, '賃料': rent, '管理費': admin_fee,
            '敷金': deposit, '礼金': key_money, '間取り': layout, '専有面積': area
        }
        properties_data.append(property_info)
    except Exception as e:
        print(f"データ抽出中にエラーが発生: {e}。この物件はスキップします。")
        continue

df = pd.DataFrame(properties_data)
print(f"合計{len(df)}件の物件情報を収集しました。")
df.head()
```
